```python
from numpy import *
from scipy.special import j1
from matplotlib.pyplot import *
import matplotlib.cm as cm

wave = 500e-9
k = 2*pi/wave
side = 1e-6
points = 400
dim = linspace(-side/2,side/2,points)
ax,ay = meshgrid(dim,dim)
r = sqrt(ax**2 + ay**2)
area = (j1(k*r)/(k*r))**2
imshow(area)
#hot()
gray()
show()
```
