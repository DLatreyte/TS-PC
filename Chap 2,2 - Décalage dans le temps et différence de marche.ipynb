# Chap. 2,2 - Décalage dans le temps et différence de marche


## Travail

Dans chacune des situations suivantes, 

1. Déterminer le décalage temporel de l'onde 2 par rapport à l'onde 1, en fonction de la période temporelle T.
2. En déduire le trajet parcouru (en plus ou en moins) par l'onde 2 par rapport à l'onde 1, en fonction de la longueur d'onde $\lambda$.
3. Donner les caractéristiques de la superposition des deux ondes (addition).

```python
import numpy as np  # Import de numpy

%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
```

```python
def wave(dates, A, T, retard=0, phase_origine=0):
    """
    Définition d'une onde scalaire
    
    Paramètres
    ----------
    dates : array
        Tableau des dates auxquelles il faut calculer l'élongation de l'onde
    A : float
        Amplitude de l'onde
    T : float
        Valeur de la période
    retard : float
        Valeur du retard
    Phase_origine : float
        Valeur de la phase à l'origine
        
    Retour
    ------
    elongations : array
        Tableau des élongations du l'onde aux dates passées en argument
    """
    retards = retard * np.ones(len(dates))
    phases_origine = phase_origine * np.ones(len(dates))
    return A * np.sin(2 * np.pi / T * (dates - retards) + phases_origine)
```

```python
A = 0.5  # Amplitude
T = 2  # Période

t_max = 3.5 * T  # dernière date
nbre_points = 1000
t = np.linspace(0, t_max, nbre_points)

```

```python
def affichage(t, onde1, onde2, onde3):
    fig = plt.figure(figsize=(10, 3.5))
    fig.subplots_adjust(left = 0.1, bottom = 0.1,
                       right = 0.9, top = 2, wspace = 0, hspace = 0.1)

    ax1 = fig.add_subplot(3, 1, 1)
    ax1.set_title(r"Première onde au point $M$")
    ax2 = fig.add_subplot(3, 1, 2)
    ax2.set_title(r"Deuxième onde au point $M$")
    ax3 = fig.add_subplot(3, 1, 3)
    ax3.set_title(r"Superposition des deux ondes au point $M$")

    ax1.set_xlabel('t (s)')
    ax2.set_xlabel('t (s)')
    ax3.set_xlabel('t (s)')

    ax1.set_ylabel('Élongation')
    ax2.set_ylabel('Élongation')
    ax3.set_ylabel('Élongation')

    ax1.set_ylim(-0.6, 0.6)
    ax2.set_ylim(-0.6, 0.6)
    ax3.set_ylim(-1.3, 1.3)

    # Choix des positions des axes de référence (4 : bas, haut, droite, gauche)
    #ax = plt.gca()  # what is the current axe
    ax1.spines['right'].set_color('none')
    ax1.spines['top'].set_color('none')
    ax1.spines['bottom'].set_position(('data', 0))
    ax1.spines['left'].set_position(('data', 0))
    ax1.xaxis.set_ticks_position('bottom')
    ax1.yaxis.set_ticks_position('left')

    ax2.spines['right'].set_color('none')
    ax2.spines['top'].set_color('none')
    ax2.spines['bottom'].set_position(('data', 0))
    ax2.spines['left'].set_position(('data', 0))
    ax2.xaxis.set_ticks_position('bottom')
    ax2.yaxis.set_ticks_position('left')

    ax3.spines['right'].set_color('none')
    ax3.spines['top'].set_color('none')
    ax3.spines['bottom'].set_position(('data', 0))
    ax3.spines['left'].set_position(('data', 0))
    ax3.xaxis.set_ticks_position('bottom')
    ax3.yaxis.set_ticks_position('left')


    ax1.plot(t, wave1)
    ax2.plot(t, wave2)
    ax3.plot(t, wave3)
```

### Première situation : Courbes en phase

```python
wave1 = wave(t, A, T)
wave2 = wave(t, A, T, 0)
wave3 = wave1 + wave2
affichage(t, wave1, wave2, wave3)
```

1. Le décalage dans le temps entre l'onde 1 et 2 est : $\Delta t = 0 + p T$ où $p$ est un entier nul ou positif. **Les deux ondes sont en phase**.
2. La différence de marche est donc $\delta = p \lambda$. L'onde 2 a parcouru une distance, en plus ou en moins par rapport à l'onde 1, égale à un multiple entier de $\lambda$.
3. L'amplitude de la somme des élongations est égale à la somme des amplitudes des deux ondes.

#### Deuxième situation : Décalage dans le temps multiple de $T/4$

```python
wave1 = wave(t, A, T)
wave2 = wave(t, A, T, T/4)
wave3 = wave1 + wave2
affichage(t, wave1, wave2, wave3)
```

1. Le décalage dans le temps entre l'onde 1 et 2 est : $\Delta t = T/4 + p T = (4p + 1) T/4 $ où $p$ est un entier nul ou positif. **Les deux ondes ne sont pas en phase**.
2. La différence de marche est donc $\delta = (4p + 1) \lambda / 4$. 
3. L'amplitude de la somme des élongations n'est égale à la somme des amplitudes des deux ondes.

#### Troisième situation : Décalage dans le temps multiple de $T/2$

```python
wave1 = wave(t, A, T)
wave2 = wave(t, A, T, T/2)
wave3 = wave1 + wave2
affichage(t, wave1, wave2, wave3)
```

1. Le décalage dans le temps entre l'onde 1 et 2 est : $\Delta t = T/2 + p T = (2p + 1) T/2 $ où $p$ est un entier nul ou positif. **Les deux ondes sont en opposition de phase**.
2. La différence de marche est donc $\delta = (2p + 1) \lambda / 2$. L'onde 2 a parcouru une distance, en plus ou en moins par rapport à l'onde 1, égale à un multiple impair de fois $\lambda / 2$.
3. L'amplitude de la somme des élongations est nulle.

#### Quatrième situation : Décalage dans le temps multiple de $3T/4$

```python
wave1 = wave(t, A, T)
wave2 = wave(t, A, T, 3*T/4)
wave3 = wave1 + wave2
affichage(t, wave1, wave2, wave3)
```

1. Le décalage dans le temps entre l'onde 1 et 2 est : $\Delta t = 3T/4 + p T = (4p + 3) T/4 $ où $p$ est un entier nul ou positif. **Les deux ondes ne sont pas en phase**.
2. La différence de marche est donc $\delta = (4p + 3) \lambda / 4$. 
3. L'amplitude de la somme des élongations n'est égale à la somme des amplitudes des deux ondes.

Cette situation est identique à la deuxième.
