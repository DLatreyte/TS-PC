```python
from matplotlib import pyplot as plt
import numpy as np
```

## Source immobile

```python
# Param√®tres
R0 = 0.5  # Rayon initial
v0 = 0.3  # Vitesse de l'onde
v1 = -0.15  # Vitesse de la source

thetas = np.linspace(0, 2*np.pi, 500)
```

```python
# Liste de cercles
nbre_cercles = 10
xs = []
ys = []

for t in range(nbre_cercles):
    x = (R0 + v0 * t) * np.cos(thetas)
    y = (R0 + v0 * t) * np.sin(thetas)
    
    xs.append(x)
    ys.append(y)
```

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)
ax.set_title("Propagation onde, source immobile")
ax.axis("Off")
bbox_props = dict(facecolor='white', boxstyle="circle")
t = ax.text(0, 0, r"$S$", ha="center", va="center", size=12, bbox=bbox_props)

for i in range(nbre_cercles):
    ax.plot(xs[i], ys[i], color="black")

fig.savefig("Effet Doppler, source immobile.pdf")
plt.show()
```

## Source en mouvement

```python
# Liste de cercles
nbre_cercles = 10
xs = []
ys = []

for t in range(nbre_cercles):
    x = v1 * t + (R0 + v0 * t) * np.cos(thetas)
    y = (R0 + v0 * t) * np.sin(thetas)
    
    xs.append(x)
    ys.append(y)
```

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)
ax.set_title("Propagation onde, source immobile")
ax.axis("Off")
bbox_props = dict(facecolor='white', boxstyle="circle")
t = ax.text(0, 0, r"$S$", ha="center", va="center", size=12, bbox=bbox_props)

for i in range(nbre_cercles):
    ax.plot(xs[i], ys[i], color="black")

fig.savefig("Effet Doppler source en mouvement.pdf")
plt.show()
```
